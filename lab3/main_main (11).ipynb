```python
import random

def generate_mas(size, low, high):
    """
    Генерирует случайные матрицы и векторы для заданного размера, исключая 0.

    Parameters:
    size (int): Размер матрицы (должен быть целым числом).
    low (int): Нижняя граница для случайных чисел.
    high (int): Верхняя граница для случайных чисел.

    Returns:
    tuple: Кортеж из шести элементов — c, b, a, p, q, x, где это случайные векторы и матрицы.
    """
    
    # Функция для генерации случайного числа без нуля
    def generate_non_zero(low, high):
        number = 0
        while number == 0:
            number = random.randint(low, high)
        return number
    
    # Генерация случайных чисел для матриц и векторов, исключая 0
    c = [generate_non_zero(low, high) for _ in range(size - 1)]
    b = [generate_non_zero(low, high) for _ in range(size)]
    a = [generate_non_zero(low, high) for _ in range(size - 1)]
    p = [generate_non_zero(low, high) for _ in range(size)]
    q = [generate_non_zero(low, high) for _ in range(size)]
    x = [random.randint(5, 10) for _ in range(size)]  # x не может быть 0, так как диапазон от 5 до 10
    
    # Преобразование элементов в тип float
    b = [float(item) for item in b]
    c = [float(item) for item in c]
    a = [float(item) for item in a]
    p = [float(item) for item in p]
    q = [float(item) for item in q]
    x = [float(item) for item in x]

    return c, b, a, p, q, x
```

```python
def do_cool(c, b, a, p, q, k, f, f_pogr):
    df = print_matrix(c, b, a, p, q, k)
    
    length = df.shape[-1] 
    
    # Проверка на деление на ноль при первой операции
    if df.iloc[0][length - 1] == 0:
        return 0
    r = 1 / df.iloc[0][length - 1]
    df.iloc[0] *= r
    f[0] *= r
    f_pogr[0] *= r

    center_left_ind = length // 2 - 1

    if k < length - 1:
        temp = -2 * (k - center_left_ind)
        
        for i in range(1, k + 1 + temp):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][length - i] == 0:
                return 0
            f[i] = f[i] - f[i - 1] * df.iloc[i][length - i]
            f_pogr[i] = f_pogr[i] - f_pogr[i - 1] * df.iloc[i][length - i]
            df.iloc[i] -= df.iloc[i - 1] * df.iloc[i][length - i]
            
            if df.iloc[i][length - i - 1] == 0:
                return 0
            f[i] /= df.iloc[i][length - i - 1]
            f_pogr[i] /= df.iloc[i][length - i - 1]
            df.iloc[i] /= df.iloc[i][length - i - 1]
            
            if i == k + temp:
                break
        
        # Проверка на деление на ноль для последней строки
        if df.iloc[length - 1][0] == 0:
            return 0
        r = 1 / df.iloc[length - 1][0]
        df.iloc[length - 1] *= r
        f[-1] *= r
        f_pogr[-1] *= r
        
        for i in range(length - 2, k + temp, -1):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][length - i - 1 - 1] == 0:
                return 0
            f[i] = f[i] - f[i + 1] * df.iloc[i][length - i - 1 - 1]
            f_pogr[i] = f_pogr[i] - f_pogr[i + 1] * df.iloc[i][length - i - 1 - 1]
            df.iloc[i] -= df.iloc[i + 1] * df.iloc[i][length - i - 1 - 1]
            
            if df.iloc[i][length - i - 1] == 0:
                return 0
            f[i] /= df.iloc[i][length - i - 1]
            f_pogr[i] /= df.iloc[i][length - i - 1]
            df.iloc[i] /= df.iloc[i][length - i - 1]
            
            if i == k + 1 + temp:
                break
        
        # Проверка на деление на ноль для строки k+temp
        if df.iloc[k + temp][k + 1] == 0:
            return 0
        r = 1 / df.iloc[k + temp][k + 1]
        df.iloc[k + temp] *= r
        f[k + temp] *= r
        f_pogr[k + temp] *= r
        
        for i in range(k + temp):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][k + 1] == 0:
                return 0
            f[i] -= f[k + temp] * df.iloc[i][k + 1]
            f_pogr[i] -= f_pogr[k + temp] * df.iloc[i][k + 1]
            df.iloc[i] -= df.iloc[i][k + 1] * df.iloc[k + temp]
        
        for i in range(k + 1 + temp, length):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][k + 1] == 0:
                return 0
            f[i] -= f[k + temp] * df.iloc[i][k + 1]
            f_pogr[i] -= f_pogr[k + temp] * df.iloc[i][k + 1]
            df.iloc[i] -= df.iloc[i][k + 1] * df.iloc[k + temp]
        
        # Проверка на деление на ноль для строки k+1+temp
        if df.iloc[k + 1 + temp][k] == 0:
            return 0
        r = 1 / df.iloc[k + 1 + temp][k]
        df.iloc[k + 1 + temp] *= r
        f[k + 1 + temp] *= r
        f_pogr[k + 1 + temp] *= r
        
        for i in range(k + 1 + temp):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][k] == 0:
                return 0
            f[i] -= f[k + 1 + temp] * df.iloc[i][k]
            f_pogr[i] -= f_pogr[k + 1 + temp] * df.iloc[i][k]
            df.iloc[i] -= df.iloc[i][k] * df.iloc[k + 1 + temp]
        
        for i in range(k + 1 + 1 + temp, length):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][k] == 0:
                return 0
            f[i] -= f[k + 1 + temp] * df.iloc[i][k]
            f_pogr[i] -= f_pogr[k + 1 + temp] * df.iloc[i][k]
            df.iloc[i] -= df.iloc[i][k] * df.iloc[k + 1 + temp]
        
        for i in range(k - 2 + temp, -1, -1):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][length - 1 - 1 - i] == 0:
                return 0
            f[i] -= f[i + 1] * df.iloc[i][length - 1 - 1 - i]
            f_pogr[i] -= f_pogr[i + 1] * df.iloc[i][length - 1 - 1 - i]
            df.iloc[i] -= df.iloc[i][length - 1 - 1 - i] * df.iloc[i + 1]

        for i in range(k + 3 + temp, length):
            # Проверки на деление на ноль для f и f_pogr
            if df.iloc[i][length - i] == 0:
                return 0
            f[i] -= f[i - 1] * df.iloc[i][length - i]
            f_pogr[i] -= f_pogr[i - 1] * df.iloc[i][length - i]
            df.iloc[i] -= df.iloc[i][length - i] * df.iloc[i - 1]
    
    return f, f_pogr
```

```python
def do_cool2(c, b, a, p, q, k, f, f_pogr):
    df = print_matrix(c, b, a, p, q, k)

    length = df.shape[-1] 
    
    # Проверка на деление на 0
    if df.iloc[0][length - 1] == 0:
        return 0
    r = 1 / df.iloc[0][length - 1]
    df.iloc[0] *= r
    f[0] *= r
    f_pogr[0] *= r

    center_left_ind = length // 2
    if k < length - 1:
        temp = -2 * (k - center_left_ind)

        for i in range(1, k + temp):
            # Проверка на деление на 0 перед операциями
            if df.iloc[i][length - i] == 0:
                return 0
            f[i] = f[i] - f[i - 1] * df.iloc[i][length - i]
            f_pogr[i] = f_pogr[i] - f_pogr[i - 1] * df.iloc[i][length - i]
            df.iloc[i] -= df.iloc[i - 1] * df.iloc[i][length - i]
            
            # Проверка на деление на 0
            if df.iloc[i][length - i - 1] == 0:
                return 0
            f[i] /= df.iloc[i][length - i - 1]
            f_pogr[i] /= df.iloc[i][length - i - 1]
            df.iloc[i] /= df.iloc[i][length - i - 1]
    
    # Проверка на деление на 0 для последней строки
    if df.iloc[length - 1][0] == 0:
        return 0
    r = 1 / df.iloc[length - 1][0]
    df.iloc[length - 1] *= r
    f[-1] *= r
    f_pogr[-1] *= r
    
    for i in range(length - 1 - 1, k - 1 + temp, -1):
        # Проверка на деление на 0
        if df.iloc[i][length - i - 1 - 1] == 0:
            return 0
        f[i] = f[i] - f[i + 1] * df.iloc[i][length - i - 1 - 1]
        f_pogr[i] = f_pogr[i] - f_pogr[i + 1] * df.iloc[i][length - i - 1 - 1]
        df.iloc[i] -= df.iloc[i + 1] * df.iloc[i][length - i - 1 - 1]
        
        # Проверка на деление на 0
        if df.iloc[i][length - i - 1] == 0:
            return 0
        f[i] /= df.iloc[i][length - i - 1]
        f_pogr[i] /= df.iloc[i][length - i - 1]
        df.iloc[i] /= df.iloc[i][length - i - 1]
    
    # Проверка на деление на 0
    if df.iloc[k - 1 + temp][k + 1] == 0:
        return 0
    r = 1 / df.iloc[k - 1 + temp][k + 1]
    df.iloc[k - 1 + temp] *= r
    f[k - 1 + temp] *= r
    f_pogr[k - 1 + temp] *= r
    
    for i in range(k - 1 + temp):
        # Проверка на деление на 0
        if df.iloc[i][k + 1] == 0:
            return 0
        f[i] -= f[k - 1 + temp] * df.iloc[i][k + 1]
        f_pogr[i] -= f_pogr[k - 1 + temp] * df.iloc[i][k + 1]
        df.iloc[i] -= df.iloc[i][k + 1] * df.iloc[k - 1 + temp]
    
    for i in range(k + temp, length):
        # Проверка на деление на 0
        if df.iloc[i][k + 1] == 0:
            return 0
        f[i] -= f[k - 1 + temp] * df.iloc[i][k + 1]
        f_pogr[i] -= f_pogr[k - 1 + temp] * df.iloc[i][k + 1]
        df.iloc[i] -= df.iloc[i][k + 1] * df.iloc[k - 1 + temp]
    
    # Проверка на деление на 0
    if df.iloc[k + temp][k] == 0:
        return 0
    r = 1 / df.iloc[k + temp][k]
    df.iloc[k + temp] *= r
    f[k + temp] *= r
    f_pogr[k + temp] *= r
    
    for i in range(k + temp):
        # Проверка на деление на 0
        if df.iloc[i][k] == 0:
            return 0
        f[i] -= f[k + temp] * df.iloc[i][k]
        f_pogr[i] -= f_pogr[k + temp] * df.iloc[i][k]
        df.iloc[i] -= df.iloc[i][k] * df.iloc[k + temp]
    
    for i in range(k + 1 + temp, length):
        # Проверка на деление на 0
        if df.iloc[i][k] == 0:
            return 0
        f[i] -= f[k + temp] * df.iloc[i][k]
        f_pogr[i] -= f_pogr[k + temp] * df.iloc[i][k]
        df.iloc[i] -= df.iloc[i][k] * df.iloc[k + temp]
    
    # Проверка на деление на 0
    for i in range(k - 1 - 1 - 1 + temp, -1, -1):
        if df.iloc[i][length - 1 - 1 - i] == 0:
            return 0
        f[i] -= f[i + 1] * df.iloc[i][length - 1 - 1 - i]
        f_pogr[i] -= f_pogr[i + 1] * df.iloc[i][length - 1 - 1 - i]
        df.iloc[i] -= df.iloc[i][length - 1 - 1 - i] * df.iloc[i + 1]
    
    for i in range(k + 1 + 1 + temp, length):
        # Проверка на деление на 0
        if df.iloc[i][length - i] == 0:
            return 0
        f[i] -= f[i - 1] * df.iloc[i][length - i]
        f_pogr[i] -= f_pogr[i - 1] * df.iloc[i][length - i]
        df.iloc[i] -= df.iloc[i][length - i] * df.iloc[i - 1]

    return df, f, f_pogr
```

```python
def otn(x, f, q_tochnost):
    delta_otn = [0 for i in range(len(f))]
    for i in range(len(f)):
        if abs(x[i]) > q_tochnost:
            delta_otn[i] = abs((f[i] - x[i])/x[i])
        else:
            delta_otn[i] = abs(f[i] - x[i])
    return max(delta_otn)
```

```python
def absol(f_pogr):
    delta_absolute = 0
    for i in range(len(f_pogr)):
            delta_absolute = max(abs(f_pogr[i]-1), delta_absolute)
    return delta_absolute
    
```
